Scenario: You are a data analyst working with a dataset of global weather. Your task is to analyze temperature trends and visualize the results.

1. Data Preparation:

    Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.

    Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

    Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.
    Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import calendar
```

```python
capital_cities = [
    "Jerusalem",    # Israel
    "Buenos Aires", # Argentina
    "Paris",        # France
    "Tokyo",        # Japan
    "London",       # United Kingdom
    "Berlin",       # Germany
    "Canberra",     # Australia
    "Ottawa",       # Canada
    "New Delhi",    # India
    "Washington D.C." # United States
]

months = list(calendar.month_name)[1:]

temperatures = [np.random.uniform(-5,35,12) for i in range(len(capital_cities))]

df_dict={}
for i in range(len(capital_cities)):
    df_dict[capital_cities[i]] = dict(zip(months, temperatures[i]))
```

I generated three lists: one containing 10 cities, one containing the 12 months of the year and one containing 12 arrays with random floats between -5 and 35 following a uniform distribution. These represent the average monthly temperatures for the 10 chosen cities. These temperatures are chosen at random and do not reflect the real temperatures of those cities at those times of the yar.

Then, I made a dictionary with all this data. For every city, there's another dictionary inside in which the keys are the months and their values, the average temperatures.

```python
df= pd.DataFrame(df_dict)
df = df.transpose()
print(df)
```

2. Data Analysis:

    Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).

    Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

    Calculate the annual average temperature for each city.
    Identify the city with the highest and lowest average temperature for the year.


```python
mean_df = df.mean(axis = 1)
print(mean_df)
```

In the table above, it is reported the average annual temperatures for 10 different cities. 

```python
print(f"The city with the highest annual temperature is {mean_df.idxmax()}.")
```

```python
print(f"The city with the lowest annual temperature is {mean_df.idxmin()}.")
```

3. Data Visualization:

```python
ax =sns.lineplot(df.transpose())
ax.set_ylabel('Temperature ºC')
ax.set_xlabel("Month")
plt.xticks(rotation= 30)
plt.tight_layout()
plt.grid()
plt.title("Monthly Temperature Trends by City")
plt.show()
```

This plot condenses all the information from the data set. Of course this doesn't reflect reality, in which cities from the southern hemisphere would show colder temperatures from april to august and hotter temperatures the rest of the year, and viceversa. All these lines should be way smoother.
Still, it is possible to distinguish the average monthly temperature from city to city.
