## Exercise 3: Building a Custom Sentiment Analyzer

Objective: Develop a more robust sentiment analyzer with direct control over components.

Instructions:

1. Import AutoTokenizer and AutoModelForSequenceClassification from transformers.
2. Create a class called BERTSentimentAnalyzer with the following methods:
  - ``__init__``:
    - Initialize the tokenizer and model with -  ‘distilbert-base-uncased-finetuned-sst-2-english’.
    - Set the device to ‘cuda’ if available, otherwise ‘cpu’.
    - Define the labels as [‘NEGATIVE’, ‘POSITIVE’].
  - ``preprocess_text``:
    - Clean the text by removing extra whitespace.
    - Tokenize the text using the BERT tokenizer with special tokens, padding, and truncation.
    - Move the tokenized inputs to the appropriate device.
  - ``predict``:
    - Preprocess the input text.
    - Get model predictions and apply softmax to obtain probabilities.
    - Return a dictionary containing the text, predicted sentiment, confidence score, and detailed probabilities.
3. Test the BERTSentimentAnalyzer with various sample texts and analyze the results.

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
```

```python
class BERTSentimentAnalyzer:
  def __init__(self, model_name='distilbert-base-uncased-finetuned-sst-2-english'):
    self.tokenizer = AutoTokenizer.from_pretrained(model_name)
    self.model = AutoModelForSequenceClassification.from_pretrained(model_name)
    self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
    self.labels = ['NEGATIVE', 'POSITIVE']
    self.model.to(self.device)

  def preprocess_text(self, text):

      text = ' '.join(text.split())

      encoding = self.tokenizer.encode_plus(
          text,
          add_special_tokens=True,
          padding='max_length',
          truncation=True,
          max_length=64,
          return_tensors='pt'
      )

      input_ids = encoding['input_ids'].to(self.device)
      attention_mask = encoding['attention_mask'].to(self.device)

      return input_ids, attention_mask

  def predict(self, text):
      input_ids, attention_mask = self.preprocess_text(text)
      with torch.no_grad():
          outputs = self.model(input_ids, attention_mask=attention_mask)
      logits = outputs.logits
      probabilities = softmax(logits, dim=1)

      predicted_class = torch.argmax(probabilities, dim=1).item()
      predicted_sentiment = self.labels[predicted_class]
      confidence_score = probabilities[0][predicted_class].item()

      result = {
          'text': text,
          'predicted_sentiment': predicted_sentiment,
          'confidence_score': confidence_score,
          'probabilities': {self.labels[i]: probabilities[0][i].item() for i in range(len(self.labels))}
      }

      return result
```

```python
analyzer = BERTSentimentAnalyzer()
```

```python
sample_text = ["I really liked the movie", "The sky is blue", "I am eating cereal", "This task is hard and I don't understand it."]
```

```python
for text in sample_text:
    result = analyzer.predict(text)
    print(f"Text: {result['text']}")
    print(f"Predicted Sentiment: {result['predicted_sentiment']}")
    print(f"Confidence Score: {result['confidence_score']:.4f}")
    print(f"Detailed Probabilities: {result['probabilities']}")
    print("-" * 50)
```

## Exercise 4: Understanding BERT for Named Entity Recognition (NER)

Objective: Explore how BERT can be used for NER tasks and understand the B-I-O tagging scheme.

Instructions:

1. Import AutoTokenizer and AutoModelForTokenClassification from transformers.
2. Create a class called BERTNamedEntityRecognizer with the following methods:
    - ``__init__``:
      - Initialize the tokenizer and model with ‘dslim/bert-base-NER’.
      - Set the device to ‘cuda’ if available, otherwise ‘cpu’.
        recognize_entities:
      - Tokenize the input text with special tokens, padding, and truncation.
      - Move the inputs to the appropriate device.
      - Get model predictions and convert them to labels.
      - Extract entities based on the B-I-O tagging scheme.
3. Test the BERTNamedEntityRecognizer with a sample text containing various named entities and analyze the output.


```python
from transformers import AutoModelForTokenClassification
class BERTNamedEntityRecognizer:
  def __init__(self, model_name='dslim/bert-base-NER'):
    self.tokenizer = AutoTokenizer.from_pretrained(model_name)
    self.model = AutoModelForTokenClassification.from_pretrained(model_name)
    self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
    self.model.to(self.device)
    self.labels = self.model.config.id2label



  def recognize_entities(self, text):
    # Tokenize the input text with padding, truncation, and special tokens
    inputs = self.tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Move input to the correct device
    input_ids = inputs['input_ids'].to(self.device)
    attention_mask = inputs['attention_mask'].to(self.device)

    # Get the model predictions (logits)
    with torch.no_grad():
        outputs = self.model(input_ids, attention_mask=attention_mask)

    # Get the predicted labels for each token
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=2).squeeze().cpu().numpy()

    # Decode tokens and associate them with labels
    tokens = self.tokenizer.convert_ids_to_tokens(input_ids.squeeze().cpu().numpy())
    predicted_labels = [self.labels[pred] for pred in predictions]

    # Extract the entities based on the B-I-O tagging scheme
    entities = []
    current_entity = None
    for token, label in zip(tokens, predicted_labels):
        if label != 'O':  # Ignore 'Outside' tokens
            if label.startswith('B-'):  # Beginning of an entity
                if current_entity:
                    entities.append(current_entity)
                current_entity = {'entity': label[2:], 'text': token}  # Label without 'B-'
            elif label.startswith('I-') and current_entity:  # Inside an entity
                current_entity['text'] += ' ' + token
        else:
            if current_entity:
                entities.append(current_entity)
                current_entity = None

    # Append the last entity if it exists
    if current_entity:
        entities.append(current_entity)

    return entities

```

```python
recognizer = BERTNamedEntityRecognizer()


sample_text = "My name is Julieta Wenger. I was born in Buenos Aires in 1996."

# Recognize entities in the sample text
entities = recognizer.recognize_entities(sample_text)
```

```python
print("Recognized Entities:")
for entity in entities:
    print(f"Entity: {entity['text']}, Type: {entity['entity']}")
```

## Exercise 6: Exploring BERT Applications in Retrieval-Augmented Generation (RAG)

Objective: Understand how BERT is used in RAG systems for information retrieval.

Instructions:

1. Research the concept of Retrieval-Augmented Generation (RAG) and its components.
2. Explain how BERT is used in the retrieval component of a RAG system.
3. Describe the role of BERT in generating embeddings for documents and queries.
4. Explain how a vector database is used to store and search for relevant documents based on BERT embeddings.
5. Provide an example of how BERT and an LLM like GPT can work together in a RAG system to answer a user’s question.


1. RAG is an architecture that combines a retriever and a generator to answer questions using external knowledge. The retriever finds relevant documents, and the generator (typically an LLM) uses them to produce a response.

2. BERT is used to create embeddings for queries and documents, enabling semantic similarity matching in the retrieval step. It helps identify relevant documents by comparing the embeddings of queries and stored texts.

3. BERT converts both queries and documents into dense vector embeddings that capture their semantic meaning. These embeddings are used for efficient similarity search.

4. A vector database stores the BERT-generated embeddings and allows fast similarity searches to retrieve the most relevant documents for a given query. It enables scalable and real-time retrieval based on vector similarity.

5. BERT retrieves the most relevant documents for a user’s question, and GPT takes those documents along with the query to generate a complete, context-aware answer. For instance, BERT finds passages on "green tea benefits," and GPT summarizes them into a fluent response.
