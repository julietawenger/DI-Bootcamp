## Exercise 5: Basic Data Analysis with Google Colab


Objective: Perform basic data analysis using Google Colab.
Task:

- Using the same notebook from the previous exercise, perform basic data analysis tasks:

  - Calculate the mean, median, and mode of a quantitative column.

  - Create a simple plot (like a histogram or bar chart) to visualize the data using Matplotlib/Seaborn libraries :

```python
print(f"Mean of the average hours per day sleeping is {round(df_americans_sleep['Avg hrs per day sleeping'].mean(),2)} hours.")
                                                       
print(f"Median of the average hours per day sleeping is {round(df_americans_sleep['Avg hrs per day sleeping'].median(),2)} hours.")    
                                                         
print(f"Mode of the average hours per day sleeping is: {df_americans_sleep['Avg hrs per day sleeping'].mode()}")                                                                                                         
```

```python
import matplotlib.pyplot as plt
```

```python
df_americans_sleep = df_americans_sleep[df_americans_sleep['Type of Days'] == "All days"]
df_americans_sleep = df_americans_sleep[df_americans_sleep['Age Group'] == "15 years and over"]
df_americans_sleep = df_americans_sleep[df_americans_sleep['Sex'] == "Both"]
plt.errorbar(df_americans_sleep['Year'], df_americans_sleep['Avg hrs per day sleeping'], yerr = df_americans_sleep['Standard Error'], fmt = '-or')
plt.xlabel("Year")
plt.ylabel("Avg hrs per day sleeping")
plt.show()
```
